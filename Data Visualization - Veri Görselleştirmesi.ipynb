```python
import pandas as pd
import numpy as np # numpy bizim numara üretmemizi, lineer cebir işlemlerini yapabildiğimiz paket.
import matplotlib.pyplot as plt  #Grafik çizen paketimizin adı.
```

```python
%matplotlib inline  
 #Jupyter' de Grafik çizimini verir. 
```

```python
x=np.arange(20) #0'dan 19'a kadar sayı üretir. 
y=np.random.normal(10,1,20) #Ortalama 10 , standart sapma 1 , 20 tane rakam
z=np.random.normal(10,2,20) #Ortalama 10 , standart sapma 2 , 20 tane rakam
#Bu satırı her çalıştırdığımda grafikte ki sayılar değişir.
```

```python
fig=plt.figure()                      #figüre tanımlıyoruz(fig)
ax=fig.add_subplot(111)               #111 anlamı 1 satır 1 grafik ve 1 sütun olacak anlamı taşır.
ax.plot(x,y,color='red')              #x'in y'ye göre değişimini gösterir. (color ile renk belirledik.)
ax.set_title('Line chart')            #Grafiğe eklediğimiz başlık
ax.set_xlabel('x')                    #x ekseninin ismini belirlemek için _xlabel() kullanılır.
ax.set_ylabel('y')                    #y ekseninin ismini belirlemek için _ylabel() kullanılır.
```

```python
fig= plt.figure()
ax=fig.add_subplot(111)  
ax.plot(x, y, label = 'Thing #1',color='red')
ax.plot(x, z, label = 'Thing #2',color='black')
ax.legend()
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(x,y, marker='o' , c='b')     #marker ekledik. Markerler iki noktanın birleştiği yeri belirli şekillere göre daha belirgin yapıyor. Burada ki c, color'ın c'si
ax.set_ylabel('$$$ Money', fontsize=10)
ax.set_xlabel('Years',fontsize=20)
fig.savefig('figure.png',dpi=600) #Bu oluşturduğumuz figürü kaydettik. Dosyalardan görebiliriz. dpi denilen şey kalite.
```

```python
fig2=plt.figure(figsize=(15,5)) #figure eni 15 inç, boyu 5 inç olacak.
a=fig2.add_subplot(131) # 1 satıra 3 grafik. O grafiklerin 1. si
a.scatter(x,y)
a.set_title('Scatter plot')  #Bir verinin diğer veriye göre noktalı şekilde gösterir.

b=fig2.add_subplot(132) # 1 satıra 3 sütun. O grafiklerin 2. si
b.hist(y)
b.set_title('Histogram')  # Verinin dağılımını gösterir.

c=fig2.add_subplot(133) # 1 satıra 3 sütun. O grafiklerin 3. si
c.boxplot([y,z])
c.set_title('Box plot')  #Outlier' ları bize gösterir.

fig2.suptitle('Visualization')
```
